# Array stacking

```python
import numpy as np
```

```python
a=np.random.randint(1,10,(2,3))
```

### Horizontal Stack

```python
a
```

```python
b=np.arange(1,10).reshape(3,3)
```

```python
b
```

```python
c=2*b
```

```python
c
```

```python
np.hstack((b,c))
```

```python
np.column_stack((b,c))
```

```python
np.concatenate((b,c),axis=1)
```

### Vertical Stack

```python
np.vstack((b,c))
```

```python
np.row_stack((b,c))
```

```python
np.concatenate((b,c),axis=0)
```

### Depth Stack

```python
np.dstack((b,c))
```
