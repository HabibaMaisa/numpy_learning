# Array Indexing and Slicing

```python
import numpy as np
```

```python
arr=np.array([1,2,3,4,5])
```

```python
print(arr)
```

```python
arr[0]
```

```python
arr[-1]
```

```python
print(arr[-1])
```

```python
arr1=np.random.randint(1,10,size=(3,3))
```

```python
print(arr1)
```

```python
print(arr1[0][1])
```

```python
print(arr1[0,1])
```

```python
arr2=np.random.randint(1,10,size=(2,3,3))
```

```python
print(arr2)
```

```python
print(arr2[0,2,2])
```

# Slicing

```python
print(arr)
```

```python
arr[0:3]
```

```python
arr[:3]
```

```python
arr[2: ]
```

```python
arr1
```

```python
arr1[:2,:2]
```
